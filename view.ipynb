```python
import os
HOME_DIR = os.environ['HOME']
```

```python
import py3Dmol
```

```python
def view_molecule(path):
    PDB_PATH = f'{HOME_DIR}/scratch/bio-out/{path}.pdb'
    # PDB_PATH = 'output_structure.pdb'
    with open(PDB_PATH, 'r') as pdb_f:
        pdb = pdb_f.read()
    view = py3Dmol.view(js='https://3dmol.org/build/3Dmol.js', width=800, height=400)
    view.addModel("".join(pdb), 'pdb')
    view.setStyle({'model': -1}, {"cartoon": {'color': 'spectrum'}})
    
    USE_01_P = True
    if USE_01_P: # probabilities for each atom
        vmin = 0.5
        vmax = 0.95
    else:
        vmin = 50
        vmax = 95
    
    view.setStyle({'cartoon': {'colorscheme': {'prop':'b','gradient': 'roygb','min': vmin,'max': vmax}}})
    return view
```

```python
view_molecule('output_structure')
```
